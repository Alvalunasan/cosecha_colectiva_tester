```python
transacciones = pd.DataFrame((cosecha_db.Transacciones * cosecha_db.CatalogoTransacciones).fetch(as_dict=True))
transacciones
```

```python

unique_socios = transacciones['Socio_id'].unique()
transaccion_socio = transacciones.groupby('Socio_id').agg(
    {'Cantidad_movimiento': [('total_transaccion', 'sum')]})
transaccion_socio.columns = transaccion_socio.columns.droplevel()

transaccion_socio_tipo = transacciones.groupby(['Socio_id', 'Tipo']).agg(
    {'Cantidad_movimiento': [('total_transaccion', 'sum')]})
transaccion_socio_tipo.columns = transaccion_socio_tipo.columns.droplevel()

transaccion_socio_tipo
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,6))
x_pos = np.linspace(1, unique_socios.shape[0], num=unique_socios.shape[0])

ax.bar(x_pos, transaccion_socio['total_transaccion'], align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('Total transaccion por socio', fontsize=16)
ax.set_xticks(x_pos)
ax.set_xticklabels(transaccion_socio.index)
ax.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
men_std = [2, 3, 4, 1, 2]
women_std = [3, 5, 2, 3, 3]
width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(x_pos, transaccion_socio['total_transaccion'], align='center', alpha=0.5, ecolor='black', capsize=10)

ax.bar(labels, men_means, width, yerr=men_std, label='Men')
ax.bar(labels, women_means, width, yerr=women_std, bottom=men_means,
       label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

plt.show()
```
